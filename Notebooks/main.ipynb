The following function gets a dataframe with our 560 songs as an input. It asks the customer to provide title and artist,after it gets audio features of this song, finds to which cluster it belongs. We also check if song belongs to Hot Song or Not Hot Song.
If it's a hot song we provide a recommendation from the same cluster and the list of hot songs
In other case, any song can be recommended but from the whole set and the same cluster


```python
def search_url(title):
    query = title
    search = sp.search(q=query, type="track")
    uri = search["tracks"]["items"][0]['external_urls']
    return uri
```

```python
def find_song(df):
    sample=pd.DataFrame()  
    df=df.copy()
       #Get name of song and artist from the client!
    fav_song = input('What is your favourite song?\n')
    print('______________________________________________________')
    singer=input('What is the artist name?\n')
    print('______________________________________________________')
    print('______________________________________________________')
    print('______________________________________________________')
    
    string = "track:" + fav_song + " artist:" + singer 
    result = sp.search(q=string, limit=5)
    #print(result)
    
    if len(result['tracks']['items']) == 0:
        print("Sorry, your song was not found!")
        sample = "No song"
        url = None
        return sample, url
    else: 
              
        song_id = result['tracks']['items'][0]['id'] #here we get the user song id
    #getting features
        features = sp.audio_features((song_id))[0]      
        dict_features = { key : [features[key]] for key in list(features.keys()) }   #making a dictionary
        features_df=pd.DataFrame(dict_features)                                #making a data frame          
        
        features_df=features_df.drop(columns = ['type', 'key', 'mode', 'id', 'uri', 'track_href',
           'analysis_url', 'duration_ms', 'time_signature'], axis=1)   #removing columns that not needed for clustering
        
    
    #scaling features, loading model and getting the user_cluster
        features_scaled_np = scaler2.transform(features_df)
        features_scaled_df = pd.DataFrame(features_scaled_np, columns=features_df.columns)
        best_model = load("../Models/gmm4_6.pickle")
        
        user_cluster = best_model.predict(features_scaled_df)[0]
     
        if (len(result['tracks']['items'])> 1):
            if song_id in list(hot_songs['id']):
                    
                      
                sample = df[(df['set']=="H") & (df['cluster_spheric']==user_cluster)].sample()
                #print("First if ",sample)
                url = search_url(sample['title'])
                print('We recommend you this song:')
            
            elif song_id in list(not_hot_songs['id']):
            
                sample = df[(df['set']=="N") & (df['cluster_spheric']==user_cluster)].sample()
                #print("elif ", sample)
                url = search_url(sample['title'])
                print('We recommend you this song:')
        
            else: 
            
                sample = df[(df['set']=="N") & (df['cluster_spheric']==user_cluster)].sample()
                #print("else ",sample)
                url = search_url(sample['title'])
                print('We recommend you this song:')
        else:
            sample = df[(df['set']=="N") & (df['cluster_spheric']==user_cluster)].sample()
                #print("else ",sample)
            url = search_url(sample['title'])
            print('We recommend you this song:')
            
           
        
        return sample, url 

       
```
